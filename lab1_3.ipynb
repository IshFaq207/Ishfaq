```python

def sum_of_squares(n):
    total = 0
    for i in range(1, n + 1):
        total += i ** 2
    return total


n = int(input("Enter a positive integer: "))

if n <= 0:
    print("Please enter a positive integer.")
else:
    result = sum_of_squares(n)
    print(f"The sum of squares from 1 to {n} is {result}")
```
