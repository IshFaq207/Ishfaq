```python

jugA, jugB = 4, 3

visited = set()

def dfs(a, b):

    if a == 2:
        print(f"Goal reached: ({a}, {b})")
        return True

    if (a, b) in visited:
        return False
    visited.add((a, b))

    print(f"Visiting: ({a}, {b})")

    if dfs(jugA, b): return True

    if dfs(a, jugB): return True

    if dfs(0, b): return True

    if dfs(a, 0): return True

    pour = min(a, jugB - b)
    if dfs(a - pour, b + pour): return True

    pour = min(b, jugA - a)
    if dfs(a + pour, b - pour): return True

    return False

print("Starting DFS from (0, 0)")
dfs(0, 0)
```
