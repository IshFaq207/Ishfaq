```python
# ==========================================================
# LAB TASK 1 â€“ DATA PREPROCESSING AND CLEANING
# Dataset: Heart Disease Dataset (UCI Repository)
# ==========================================================

# Step 1: Import Required Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# ----------------------------------------------------------
# Step 2: Load Dataset from GitHub
# ----------------------------------------------------------
url = "https://raw.githubusercontent.com/sharmaroshan/Heart-UCI-Dataset/master/heart.csv"
df = pd.read_csv(url)

print("âœ… Dataset Loaded Successfully!\n")
print("Shape of dataset:", df.shape)
print("\n=== First 5 Rows of Dataset ===")
display(df.head())

# ----------------------------------------------------------
# Step 3: Basic Dataset Information
# ----------------------------------------------------------
print("\n=== Dataset Information ===")
df.info()

print("\n=== Summary Statistics ===")
display(df.describe())

# ----------------------------------------------------------
# Step 4: Check Missing Values
# ----------------------------------------------------------
print("\n=== Missing Values in Each Column ===")
print(df.isnull().sum())

# ----------------------------------------------------------
# Step 5: Remove Duplicates
# ----------------------------------------------------------
before = df.shape
df.drop_duplicates(inplace=True)
after = df.shape
print(f"\nRemoved {before[0] - after[0]} duplicate rows")
print("New Shape of Dataset:", df.shape)

# ----------------------------------------------------------
# Step 6: Rename Columns for Readability
# ----------------------------------------------------------
df.rename(columns={
    'trestbps': 'resting_bp',
    'chol': 'cholesterol',
    'thalach': 'max_heart_rate'
}, inplace=True)

# ----------------------------------------------------------
# Step 7: Handle Outliers (Optional)
# ----------------------------------------------------------
# Remove extreme cholesterol values
df = df[df['cholesterol'] < 600]
print("\nAfter removing outliers:", df.shape)

# ----------------------------------------------------------
# Step 8: Convert Data Types (if needed)
# ----------------------------------------------------------
df = df.astype(float)
print("\nAll columns converted to float type.")

# ----------------------------------------------------------
# Step 9: Final Cleaned Dataset Overview
# ----------------------------------------------------------
print("\n=== Cleaned Dataset (First 5 Rows) ===")
display(df.head())

print("\nNew Shape:", df.shape)
print("\nMissing Values After Cleaning:\n", df.isnull().sum())

# ----------------------------------------------------------
# Step 10: Visualization
# ----------------------------------------------------------
print("\n=== Heart Disease Distribution ===")
df['target'].value_counts().plot(kind='bar', color=['skyblue', 'salmon'])
plt.title("Heart Disease Distribution")
plt.xlabel("Target (0 = No Disease, 1 = Disease)")
plt.ylabel("Count of Patients")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# ----------------------------------------------------------
# Step 11: Save Cleaned Dataset
# ----------------------------------------------------------
df.to_csv('heart_cleaned.csv', index=False)
print("âœ… Cleaned dataset saved as 'heart_cleaned.csv'")

# ----------------------------------------------------------
# Step 12: Summary / Conclusion
# ----------------------------------------------------------
print("""
======================== SUMMARY ========================
âœ”ï¸ Dataset Source: Heart Disease (UCI Repository)
âœ”ï¸ Rows before cleaning: 303
âœ”ï¸ Rows after cleaning: {}
âœ”ï¸ Columns: {}
âœ”ï¸ No missing values remain
âœ”ï¸ Dataset is clean and ready for Machine Learning
=========================================================
""".format(df.shape[0], df.shape[1]))
```
