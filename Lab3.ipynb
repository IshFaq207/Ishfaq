```python
# Depth First Search (DFS) implementation in Python

# Define the graph as a dictionary
graph = {
    'A': ['B'],
    'B': ['C', 'D'],
    'C': ['E'],
    'D': [],
    'E': ['F'],
    'F': []
}

# DFS function
def dfs(graph, start, goal, path=None, visited=None):
    if path is None:
        path = []
    if visited is None:
        visited = set()

    path.append(start)
    visited.add(start)

    # If goal node found
    if start == goal:
        return path

    # Explore neighbors
    for neighbor in graph[start]:
        if neighbor not in visited:
            result = dfs(graph, neighbor, goal, path.copy(), visited.copy())
            if result is not None:
                return result

    return None

# Input values
start_node = 'A'
goal_node = 'F'

# Perform DFS
path = dfs(graph, start_node, goal_node)

# Print the result
if path:
    print("DFS Path from", start_node, "to", goal_node, "is:", ' → '.join(path))
else:
    print("No path found from", start_node, "to", goal_node)
```
