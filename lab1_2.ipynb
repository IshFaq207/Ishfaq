```python

def is_prime(num):

    if num <= 1:
        return False

    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True


num = int(input("Enter a positive integer (0 - 1000): "))


if num < 0 or num > 1000:
    print("Please enter a number between 0 and 1000.")
else:

    if is_prime(num):
        print("Prime number")
    else:
        print("Not a prime number")
```
